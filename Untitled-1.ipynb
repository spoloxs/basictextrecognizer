```python
import numpy as np
from skimage import io
import os
from PIL import Image
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
```

```python
def add_margin(pil_img, top, right, bottom, left, color):
  width, height = pil_img.size
  new_width = width + right + left
  new_height = height + top + bottom
  result = Image.new(pil_img.mode, (new_width,   new_height), color)
  result.paste(pil_img, (left, top))
  return result
```

```python
dataset = []
labels = []
count = 0
folders = os.listdir(r"./English/Fnt/")

for i in folders:
    for j in os.listdir(r"./English/Fnt/" + str(i)):
        im_new = add_margin(Image.open(r"./English/Fnt/" + str(i) + '/' + str(j)).convert('L'), 10, 10, 10, 10, (255))
        resized_image = im_new.resize((64, 64))
        dataset.append(np.array(resized_image) / 255.0)
        if count <= 9:
            labels.append(count)
        elif 10 <= count <= 35:
            labels.append(count)
        else:
            labels.append(count)
    count += 1
```

```python
dataset = np.array(dataset)
labels = np.array(labels)
```

```python
model = Sequential([
    Conv2D(32, (3, 3), input_shape=(64, 64, 1), activation='relu'),
    Conv2D(32, (3, 3), activation='relu'),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dense(62, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(dataset, labels, test_size=0.2, random_state=42)
```

```python
model.fit(x_train, y_train, epochs=6, validation_data=(x_val, y_val))
```

```python
import cv2
import numpy as np

# Function to perform text detection using OpenCV
def detect_text(image_path):
    # Read the image
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Apply thresholding
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Find contours
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Filter contours based on area
    min_area = 50
    max_area = 10000
    filtered_contours = [cnt for cnt in contours if min_area < cv2.contourArea(cnt) < max_area]

    # Extract bounding boxes of text regions
    text_boxes = [cv2.boundingRect(cnt) for cnt in filtered_contours]
    
    return text_boxes
```

```python
# Path to the paragraph image
paragraph_image_path = '/home/laptap/Pictures/screenshot_2024-05-30_03-40-32.png'

# Detect text regions in the paragraph image
text_boxes = detect_text(paragraph_image_path)

# Draw bounding boxes around detected text regions
image = cv2.imread(paragraph_image_path)
for (x, y, w, h) in text_boxes:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

cv2.imshow('Detected Text', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def recognize_text(image, text_boxes):
    extracted_text = []
    character_images = []
    for (x, y, w, h) in text_boxes:
        # Crop the region of interest (ROI) defined by the bounding box
        roi = image[y:y+h, x:x+w]
        
        # Resize the ROI to fit the model input shape
        roi_resized = cv2.resize(roi, (64, 64))
        
        # Convert the ROI to grayscale and normalize
        roi_normalized = cv2.cvtColor(roi_resized, cv2.COLOR_BGR2GRAY) / 255.0
        
        # Expand the dimensions to match the model input shape
        roi_expanded = np.expand_dims(roi_normalized, axis=0)
        roi_expanded = np.expand_dims(roi_expanded, axis=-1)
        
        # Predict the character using the model
        prediction = model.predict(roi_expanded)
        predicted_class = np.argmax(prediction, axis=1)
        predicted_character = label_to_character[predicted_class[0]]
        
        # Append the predicted character to the list
        extracted_text.append(predicted_character)
        character_images.append(roi_resized)
    
    return extracted_text, character_images
```

```python
# Detect text regions in the paragraph image
text_boxes = detect_text(paragraph_image_path)

# Read the image
image = cv2.imread(paragraph_image_path)

# Extract text within the bounding boxes using the trained model
extracted_text, character_images = recognize_text(image, text_boxes)

# Show each character bounded by a box as an image
for i, (text, char_img) in enumerate(zip(extracted_text, character_images)):
    print(text)
    cv2.imshow(f'Character {i + 1}: {text}', char_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
